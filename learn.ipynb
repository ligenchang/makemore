People use the log of the likelihood (log-likelihood) instead of the raw likelihood for several key reasons:

 1. Numerical Stability
Likelihoods are products of many probabilities, and probabilities are often small (e.g., 0.01 or 0.001). Multiplying many small numbers quickly leads to underflow, where the result becomes too small for the computer to represent accurately.

Using logs turns this:


P1 * P2 * P3 * ... * Pn
into:

log(P1) + log(P2) + log(P3) + ... + log(Pn)
which is much safer numerically.

 2. Simplifies Derivatives (for Optimization)
In machine learning and statistics, we often maximize likelihoods (or minimize negative log-likelihoods). Taking the log simplifies the math:

The log of a product becomes a sum.

Derivatives of sums are much easier to work with than products.

 3. Better Interpretability
The log-likelihood gives a more interpretable sense of model fit, especially when comparing different models. A higher log-likelihood means a better fit to the data.

 4. Standard Practice in Loss Functions
In models like language models, logistic regression, etc., we minimize Negative Log-Likelihood (NLL), which becomes a natural loss function.

log (a * b * c) = log(a) + log(b) + log(c)


```python
log_likelihood = 0.0  # Initialize the log-likelihood to zero
n = 0  # Counter for the total number of bigrams processed

for w in words[:3]:  # Iterate over the first 3 words in the dataset
    chs = ['.'] + list(w) + ['.']  # Add start ('.') and end ('.') tokens to the word
    for ch1, ch2 in zip(chs, chs[1:]):  # Iterate over consecutive character pairs (bigrams)
        ix1 = stoi[ch1]  # Map the first character of the bigram to its index
        ix2 = stoi[ch2]  # Map the second character of the bigram to its index

        prob = P[ix1, ix2]  # Retrieve the probability of the bigram from the normalized tensor `P`
        logprob = torch.log(prob)  # Compute the log of the probability (logarithmic scale for stability)

        log_likelihood += logprob  # Accumulate the log-likelihood
        n += 1  # Increment the bigram counter

        # Print the bigram, its probability, and its log-probability
        print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')

# Print the total log-likelihood
print(f'{log_likelihood=}')

nll = -log_likelihood  # Compute the negative log-likelihood (NLL), which is the loss function
print(f'{nll=}')

# Print the average negative log-likelihood per bigram (useful for model evaluation)
print(f'{nll/n}')
```

```python
xenc = F.one_hot(xs, num_classes=27).float() # input to the network: one-hot encoding
logits = xenc @ W # predict log-counts, this is just multipliation and addition

#these two steps are softmax    
counts = logits.exp() # counts, equivalent to N. exponentiating the logits, we how know to propagate through
probs = counts / counts.sum(1, keepdims=True) # probabilities for next character
# btw: the last 2 lines here are together called a 'softmax'
```

 Goal:
You're trying to predict the next character (represented as an integer) in a sequence. Think of this like a tiny language model  it looks at one character and tries to guess what comes next.

 Step-by-step Intuition:
1. Your training data:
python
Copy
Edit
xs = tensor([ 0,  5, 13, 13,  1])
ys = tensor([ 5, 13, 13,  1,  0])
These are like mini text sequences:

When you see a 0, the next character is 5.

When you see a 5, the next is 13.

When you see a 13, the next is again 13, and so on.

2. One-hot encoding the input:
python
Copy
Edit
xenc = F.one_hot(xs, num_classes=27).float()
Now instead of just numbers like 0, 5, 13, you're turning each into a 27-dimensional vector with a 1 in the corresponding position. So:

0 becomes [1, 0, 0, ..., 0]

5 becomes [0, 0, 0, 0, 0, 1, 0, ..., 0]

etc.

You're telling the model: "Hey, this is the current character I'm seeing."

3. Weights matrix W:
python
Copy
Edit
W = torch.randn((27, 27), generator=g)
Each of the 27 neurons corresponds to one possible next character. They each look at the 27 one-hot features from the input and output a score (logit) for how likely that character is to come next.

Think of W as:

A table where each row is a neuron for a possible next character.

Each column is a weight corresponding to a position in the one-hot input.

So you're doing:

python
Copy
Edit
logits = xenc @ W
which is just:

"Take the one-hot input and see how strongly it activates each of the 27 neurons."

4. From logits to probabilities (the "softmax"):
python
Copy
Edit
counts = logits.exp()
probs = counts / counts.sum(1, keepdims=True)
The logits are unnormalized scores  they could be negative, positive, big, small.

.exp() turns them into positive counts  like raw vote counts.

Then you normalize them by dividing by the total sum:

"What fraction of the total vote did each possible character get?"

This is softmax: it turns raw scores into a probability distribution over the next character.

 Summary
Youre doing a super basic neural network:

Input: One-hot vector (what character were looking at)

Weights: One set of weights per possible next character

Output: Probabilities of each possible next character, computed using softmax.
